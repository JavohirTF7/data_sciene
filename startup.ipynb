```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```
startup=pd.read_csv("/content/drive/MyDrive/startups.csv")
startup.head(15)
```

```
taxonomy=pd.read_csv("/content/drive/MyDrive/taxonomy.csv")
taxonomy.head()
```

```
taxonomy.shape
```

```
taxonomy.isnull().sum()
```

```
tran=pd.read_csv("/content/drive/MyDrive/transactions.csv")
tran.head(10)
```

```
tran.shape
```

```
tran.isnull().sum()
```

```
tran.info()
```

```
dfff=pd.merge(startup,tran)
dfff
```

```
df=pd.merge(dfff,taxonomy)
df.head(30)
```

```
df.isnull().sum()
```

```
df.info()
```

```
df['last_stage_known'].unique()
```

```
import datetime as dt
```

```
df['date_announced'].str.len().value_counts()
```

```
df['date_announ']=pd.to_datetime(df['date_announced'],format="%Y-%m-%d")   # object to datetime format
```

```
df.info()
```

```
df.head()
```

```
df['year_announced']=df['date_announ'].dt.year
```

```
df.head()
```

```
df=df.drop('date_announced',axis=1)   #date_annonced no need, so deleted
```

```
df['diff_year']=df['year_announced']-df['year_founded']  # year_announced-year_founded 
```

```
df.head()
```

```
df.isnull().sum()
```

## NaN values will delete

```
df=df.dropna(subset=['year_founded'])  #year_founded's NaN values deleted
```

```
df.isnull().sum()
```

```
df.head()
```

```
df['diff_year'].value_counts()
```

```
df=df[df['diff_year']>0]
```

```
df.hist(bins=50,figsize=(20,15))
plt.show()
```
